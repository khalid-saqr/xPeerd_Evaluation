# TRACE-R reproducibility
Public Colab launcher for the version-pinned Zenodo analysis.


```python
# CELL 1 — Install repository package
import os, subprocess, sys
REPO='https://github.com/khalid-saqr/xPeerd_Evaluation.git'
ROOT='/content/xPeerd_Evaluation'
if not os.path.isdir(ROOT): subprocess.check_call(['git','clone','--depth','1',REPO,ROOT])
os.chdir(ROOT+'/TRACE_R_reproducibility')
subprocess.check_call([sys.executable,'-m','pip','install','-q','-e','.'])
print('TRACE-R installed')
```

```python
# CELL 2 — Run the locked reproduction
from pathlib import Path
from trace_r.acquisition import download, FILENAME
from trace_r.pipeline import run_analysis
WORK=Path('/content/TRACE-R-reproduction'); WORK.mkdir(exist_ok=True)
source=download(WORK/FILENAME)
report=run_analysis(source,WORK/'outputs')
report
```

```python
# CELL 3 — Verify completion status
assert report['status']=='PASS'
assert report['quality_checks_passed']==report['quality_checks_total']==22
print(report)
```

```python
# CELL 4 — Inspect transparent exclusions
import pandas as pd
exclusions=pd.read_csv(WORK/'outputs/tables/exclusion_summary.csv')
exclusions
```

```python
# CELL 5 — Inspect paired TRACE-R source profile
profile=pd.read_csv(WORK/'outputs/tables/trace_r_source_profile.csv')
profile.head()
```

```python
# CELL 6 — Inspect correlation and agreement tests
corr=pd.read_csv(WORK/'outputs/tables/cross_source_correlation_and_agreement_tests.csv')
corr
```

```python
# CELL 7 — Inspect category correspondence
categories=pd.read_csv(WORK/'outputs/tables/category_paired_correspondence_tests.csv')
categories
```

```python
# CELL 8 — Inspect recommendation agreement
recommendations=pd.read_csv(WORK/'outputs/tables/recommendation_consensus_tests.csv')
recommendations
```

```python
# CELL 9 — Compare regenerated tables with reference hashes
subprocess.check_call([sys.executable,'compare_reference.py',str(WORK/'outputs')])
```

```python
# CELL 10 — Package all regenerated evidence
import shutil
archive=shutil.make_archive('/content/TRACE_R_outputs','zip',root_dir=WORK/'outputs')
print(archive)
```
